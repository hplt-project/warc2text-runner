```python
import pandas as pd
```

# CLD2 langauge name

```python
# Info on languge abbreviations from CLD2 repo
cld2_langs = 'https://raw.githubusercontent.com/bitextor/cld2/30721661694d09625f57d3eb0a6d0ee733f0b57f/docs/evaluate_cld2_large_20140122.txt'
linf=pd.read_csv(cld2_langs, skiprows=2, sep='\t')
linf = linf.rename(columns={' Name':'Name'})
linf.Name = linf.Name.str.lower()
linf
```

```python
linf = linf[['Name','Code','Script']].groupby('Code').agg(lambda l: ' / '.join(set(l)))
linf.query('Code=="uz"')
```

```python
linf.reset_index()[['Code','Name']].to_csv('./_langs/code2name_cld2.tsv', sep='\t', index=False)
```
